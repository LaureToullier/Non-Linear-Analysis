```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
import math
import pandas as pd
```

```python
linear7 = pd.read_csv("107_linear.csv")
nonlinear7 = pd.read_csv("Non_lin_load_10^7.csv")
linear8 = pd.read_csv("108_linear.csv")
nonlinear8 = pd.read_csv("Non_lin_load_10^8.csv")
linear9 = pd.read_csv("109_linear.csv")
#nonlinear9 = pd.read_csv("Non_lin_load_10^9.csv")
```

```python
def printgraph(x,y,xname,yname,unit1, unit2, **kwargs):
    plt.plot(x,y,**kwargs)
    
    # Show a straight line
    plt.axline((x[0], y[0]), (x[1], y[1]))

    plt.title(f'{yname} as a function of {xname}')
    plt.xlabel(f"{xname} ({unit1})")
    plt.ylabel(f"{yname} ({unit2})")

    return None
```

```python
plt.figure(figsize=(7, 5))
printgraph(linear7.u1,linear7.p1,"u1","p1","mm","N", label="geometric linearity", marker="o")
printgraph(nonlinear7.u1,nonlinear7.p1,"u1","p1","mm","N", label="geometric non linearity", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(linear7.u2,linear7.p2,"u2","p2","mm","N",label="geometric linearity", marker="o")
printgraph(nonlinear7.u2,nonlinear7.p2,"u2","p2","mm","N",label="geometric non linearity", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(linear7.u1,linear7.u2,"u1","u2","mm","mm",label="geometric linearity", marker="o")
printgraph(nonlinear7.u1,nonlinear7.u2,"u1","u2","mm","mm",label="geometric non linearity", marker="o")
plt.legend()
plt.show()
```
