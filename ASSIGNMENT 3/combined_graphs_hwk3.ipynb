```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
import math
import pandas as pd
```

```python
lin107 = pd.read_csv("107_linear.csv")
lin108 = pd.read_csv("108_linear.csv")
lin109 = pd.read_csv("109_linear.csv")

nonlin107 = pd.read_csv("Non_lin_load_10^7.csv")
nonlin108 = pd.read_csv("Non_lin_load_10^8.csv")
nonlin109 = pd.read_csv("Non_lin_load_10^9.csv")

print(np.max(lin107.p1), np.max(nonlin107.p1))
print(np.max(lin107.p2), np.max(nonlin107.p2))
print(np.max(lin107.u1), np.max(nonlin107.u1))
print(np.max(lin107.u2), np.max(nonlin107.u2))

print(" ")

print(np.max(lin108.p1), np.max(nonlin108.p1))
print(np.max(lin108.p2), np.max(nonlin108.p2))
print(np.max(lin108.u1), np.max(nonlin108.u1))
print(np.max(lin108.u2), np.max(nonlin108.u2))

print(" ")

print(np.max(lin109.p1), np.max(nonlin109.p1))
print(np.max(lin109.p2), np.max(nonlin109.p2))
print(np.max(lin109.u1), np.max(nonlin109.u1))
print(np.max(lin109.u2), np.max(nonlin109.u2))
```

```python
def printgraph(x,y,xname,yname,unit1, unit2, **kwargs):
    plt.plot(x,y,**kwargs)
    
    # Show a straight line
    plt.axline((x[0], y[0]), (x[1], y[1]))

    plt.title(f'{yname} as a function of {xname}')
    plt.xlabel(f"{xname} ({unit1})")
    plt.ylabel(f"{yname} ({unit2})")

    return None
```

```python
def percents(lin, nonlin, applied_force):
    displacements = ["u1", "u2"]
    for axis in displacements:

        diffabs = abs(np.max(nonlin[axis]) - np.max(lin[axis]))
        average =  np.max(nonlin[axis]) + np.max(lin[axis])
        percentchange = diffabs / average * 100

        diff = np.max(nonlin[axis]) - np.max(lin[axis])
        percent = diff / np.max(lin[axis]) * 100


        print(f"Max displacement percentage difference between non-linear and linear geometries with {applied_force} N force in {axis} direction: {percent} %")

    return None

percents(lin107, nonlin107, "10^7")
percents(lin108, nonlin108, "10^8")
percents(lin109, nonlin109, "10^9")
```

```python
# 10^7

plt.figure(figsize=(7, 5))
printgraph(lin107.u1,lin107.p1,"u1","p1","mm","N", label="Geometric linearity", marker="o")
printgraph(nonlin107.u1,nonlin107.p1,"u1","p1","mm","N", label="Geometric non-linearity", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(lin107.u2,lin107.p2,"u2","p2","mm","N", label="Geometric linearity", marker="o")
printgraph(nonlin107.u2,nonlin107.p2,"u2","p2","mm","N", label="Geometric non-linearity", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(lin107.u1,lin107.u2,"u1","u2","mm","N", label="Geometric linearity", marker="o")
printgraph(nonlin107.u1,nonlin107.u2,"u1","u2","mm","N", label="Geometric non-linearity", marker="o")
plt.legend()
plt.show()
```
