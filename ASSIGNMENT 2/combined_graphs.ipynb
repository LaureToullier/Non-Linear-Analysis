```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
import math
import pandas as pd
```

```python
partA = pd.read_csv("partA.csv")
partB = pd.read_csv("partB.csv")
```

```python
def printgraph(x,y,xname,yname,unit1, unit2, **kwargs):
    plt.plot(x,y,**kwargs)
    
    # Show a straight line
    plt.axline((x[0], y[0]), (x[1], y[1]))

    plt.title(f'{yname} as a function of {xname}')
    plt.xlabel(f"{xname} ({unit1})")
    plt.ylabel(f"{yname} ({unit2})")

    return None
```

```python
plt.figure(figsize=(7, 5))
printgraph(partA.u1,partA.p1,"u1","p1","mm","N", label="Event-to-event method", marker="o")
printgraph(partB.u1,partB.p1,"u1","p1","mm","N", label="Newton Raphson method", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(partA.u2,partA.p2,"u2","p2","mm","N",label="Event-to-event method", marker="o")
printgraph(partB.u2,partB.p2,"u2","p2","mm","N",label="Newton Raphson method", marker="o")
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
printgraph(partA.u1,partA.u2,"u1","u2","mm","mm",label="Event-to-event method", marker="o")
printgraph(partB.u1,partB.u2,"u1","u2","mm","mm",label="Newton Raphson method", marker="o")
plt.legend()
plt.show()
```
